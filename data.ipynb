```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("All_India_Index_july2019_20Aug2020_dec20_1_1.csv")
```

# Scatter Plot

```python
column = list(data.columns)
column = column[3:9]

jan = list(data.loc[data['Month'] == "January"].iloc[0])[3:9]
july = list(data.loc[data['Month'] == "July"].iloc[0])[3:9]
dec = list(data.loc[data['Month'] == "December"].iloc[0])[3:9]

plt.figure(figsize=(15, 10))
plt.scatter(column,jan,label="January")
plt.scatter(column,july,label="July")
plt.scatter(column,dec,label="December")
plt.title("All India Consumer Price Index of Various Products for January, July and December of 2013", fontsize=21)
plt.ylabel("All India Consumer Price Index", fontsize=25)
plt.xlabel("Products", fontsize=25)
plt.xticks(rotation = 60)
plt.tick_params(axis='both', labelsize=15)
plt.legend(fontsize=25)
# plt.show()
plt.savefig("figures/scatter.png")
```

# Box Plot

```python
products = []
for i in column:
    dataProducts = data[i].iloc[:36].to_numpy()
    products.append(dataProducts)
    
plt.figure(figsize=(15, 10))
plt.boxplot(products,labels=column, patch_artist=True)
plt.title("All India Consumer Price Index of Various Products for 2013", fontsize=25)
plt.ylabel("All India Consumer Price Index", fontsize=25)
plt.xlabel("Products", fontsize=25)
plt.xticks(rotation = 60)
plt.tick_params(axis='both', labelsize=15)
# plt.show()
plt.savefig("figures/box.png")
```

# Bar Plot

```python
plt.figure(figsize=(20, 10))
plt.bar(data["Month"].iloc[:36], data["Education"].iloc[:36])
plt.title("Month-wise All India Consumer Price Index of Education for 2013", fontsize=30)
plt.ylabel("All India Consumer Price Index", fontsize=25)
plt.xlabel("Months",fontsize=25)
plt.xticks(rotation = 60)
plt.tick_params(axis='both', labelsize=15)
plt.ylim([100, 115])
# plt.show()
plt.savefig("figures/bar.png")
```
