```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("All_India_Index_july2019_20Aug2020_dec20_1_1.csv")
```

Scatter Plot

```python
colm = list(data.columns)
colm = colm[3:12]

jan = list(data.loc[data['Month'] == "January"].iloc[0])[3:12]
july = list(data.loc[data['Month'] == "July"].iloc[0])[3:12]
dec = list(data.loc[data['Month'] == "December"].iloc[0])[3:12]

plt.figure(figsize=(15, 10))
plt.scatter(colm,jan,label="January")
plt.scatter(colm,july,label="July")
plt.scatter(colm,dec,label="December")
plt.title("Month wise Price Index of Various Products", fontsize=25)
plt.ylabel("Price Index", fontsize=25)
plt.xlabel("Months", fontsize=25)
plt.xticks(rotation = 60)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.legend(fontsize=25)
plt.show()
```

Box Plot

```python
monthwise = []
for year in colm:
    df_year = data[year].iloc[:36].to_numpy()
    monthwise.append(df_year)
    
plt.figure(figsize=(15, 10))
plt.boxplot(monthwise,labels=colm, patch_artist=True)
plt.title("Month wise Price Index of Various Products", fontsize=25)
plt.ylabel("Price Index", fontsize=25)
plt.xlabel("Products", fontsize=25)
plt.xticks(rotation = 60)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.show()
```

Bar Plot

```python
plt.figure(figsize=(20, 10))
plt.bar(data["Month"].iloc[:36], data["Education"].iloc[:36])
plt.title("Month-wise All India Consumer Price Index of Education for 2013", fontsize=30)
plt.ylabel("Consumer Price Index", fontsize=25)
plt.xlabel("Months",fontsize=25)
plt.xticks(rotation = 60)
plt.tick_params(axis='both', which='major', labelsize=15)
plt.ylim([95, 115])
plt.show()
```
